# urllib包
## request类 
- urlopen() 函数 负责请求网页
    格式： urllib2.urlopen(url, data, timeout, cafile, capath, cadefault, context)
    timeout 指定了一个以秒为单位的超时参数来屏蔽比如尝试链接的操作(如果没有设定，默认的全局超市设置将会被采用)。 这个参数只有对             HTTP，HTTPS和 FTP 链接起作用   
         
- read（）函数  负责读取请求到的网页
        格式： rsp.read(）
         
- decode（）函数 进行解码 
        格式：html.decode() 参数为“编码格式”，可以手动输入编码格式，也可用第三方包：chardet
        
- date()
    - 访问网络的两种方法： get和 post
        - get：
            - 利用参数给服务器传递信息
            - 参数为字典结构 需用parse编码
        - post：
            - 一般向服务器传递参数使用
            - psot是把信息自动加密
            - 如果用post信息，需要使用date参数
            - 使用post，意味着Http的请求头需要更改
                -Content-Type: application/x-www.for-urlencode
                -Content-Length: 数据长度
            - urllib.parse.urlencode可以把字符串自动传唤
    - 若要使用date（），需使用字典结构

```python
from urllib import parse
if __name__ == '__main__':
    url = "http://www.baidu.com/s?"
    wd = input("Input your keyword:")
    
    qs = {
        "wd" : wd
    }
    # 对输入字符qs 进行编码
    qs = parse.urlencode(qs)
    
    fullurl = url + qs
    print(fullurl)
    rsp = request.urlopen(fullurl).read().decode()
    print(rsp)
```

```python
# 测试 encode()
di = {
    'kw': '无'
}
di_code = parse.urlencode(di)
print(di_code)
di_code = di_code.encode()
print(di_code)
```

```python
# 测试error
from urllib import request, error

if __name__ == '__main__':
    #网址错误
    url = "http://www.caidu.com"
    try:
        req = request.Request(url)
        rsq = request.urlopen(req)
        html = rsq.read().decode()
        print(html)
        
    except error.URLError as e:
        #打印错误原因
        print("URLError: {0}".format(e.reason))
        print("URLError:{0}".format(e))
    except Exception as e:
        print(e)
  
```

# UserAgent（用户代理）
- UserAgent： 用户代理，简称UA，属于heads的一部分，服务器通过UA判断访问者身份
- 常见UA值，使用的时候可以直接复制，也可以通过chrome访问时抓包
- 设置UA的方式;
    - heads
    - add_header

```python
# UA测试
from urllib import request,parse,error
if __name__ == "__main__":
    url = "http://www.baidu.com"
    try:
        #heads
        headers={}
        headers['User-Agent']= 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36'
        #add_header('User-Agent','Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/70.0.3538.110 Safari/537.36')
        req = request.Request(url,headers=headers)
        rsp = request.urlopen(req)
        html = rsp.read().decode()
        print(html)
    except URLError as u:
        print("URLRrror:{0}".format(u))
```
